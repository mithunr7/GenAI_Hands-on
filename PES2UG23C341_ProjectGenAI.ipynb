```python
# !pip install transformers torch

from transformers import pipeline

# Load toxicity detection model
print("Loading model...")
classifier = pipeline('text-classification', model='unitary/toxic-bert')
print("âœ… Ready!\n")

def check_toxicity(text):
    """Check if text is toxic/bullying"""
    result = classifier(text)[0]
    
    is_toxic = result['label'] == 'toxic'
    confidence = result['score']
    
    return {
        'text': text,
        'is_toxic': is_toxic,
        'confidence': confidence
    }

# Test examples
print("="*60)
print("ðŸ›¡ï¸ CYBERBULLYING DETECTOR")
print("="*60)

# Test comments
comments = [
    "You're so stupid and ugly!",
    "Great job on your project!",
    "Nobody likes you, loser",
    "I love your work, keep it up!",
    "You should just quit, you're terrible"
]

for comment in comments:
    result = check_toxicity(comment)
    
    status = "ðŸš¨ TOXIC" if result['is_toxic'] else "âœ… SAFE"
    conf = result['confidence'] * 100
    
    print(f"\n{status} ({conf:.1f}%)")
    print(f"Comment: \"{result['text']}\"")

# Interactive mode
print("\n" + "="*60)

def interactive():
    while True:
        text = input("\nEnter comment to check (or 'quit'): ").strip()
        
        if text.lower() == 'quit':
            break
        
        if not text:
            continue
        
        result = check_toxicity(text)
        
        if result['is_toxic']:
            print(f"ðŸš¨ TOXIC - Confidence: {result['confidence']*100:.1f}%")
            print("âš ï¸ This comment may contain bullying/harassment")
        else:
            print(f"âœ… SAFE - Confidence: {result['confidence']*100:.1f}%")

# Uncomment to use:
interactive()
```
